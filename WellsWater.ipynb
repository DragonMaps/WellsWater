```python
WellsInCounties = gpd.sjoin(wells_gdf,co_counties_gdf.to_crs(wells_gdf.crs), how='inner', predicate='intersects')

WellsInCounties['Wells']=1

WellsInCounties.head(10)
```

```python
WellsInCounties.to_crs(co_counties_gdf.crs)

WellsInCounties = WellsInCounties.groupby('LABEL').sum()

WellsInCounties
```

```python
#Get #wells for counties which have wells. 

Wells = WellsInCounties['Wells'].tolist()

for i in range(len(Wells)):
#    VotersInPrecincts.index[i],Voters[i])
    county = WellsInCounties.index[i]
    
    county_index = co_counties_gdf[(co_counties_gdf['LABEL'])==county].index[:]
#    print('JeffCo Index:',len(jeffco_index),jeffco_index)
#    print(prec,jeffco_precincts_gdf[(jeffco_precincts_gdf['COUNTYPREC'])==prec])
    for j in range(0,len(county_index)):
#        jeffco_precincts_gdf['Voters'][jeffco_index[j]]=Voters[i]
        co_counties_gdf.loc[county_index[j],'Wells']=Wells[i]
#        print(prec,j,Voters[i],jeffco_precincts_gdf['Voters'][j],jeffco_index)
#        print(VotersInprecincts.index[i],jeffco_index,jeffco_precincts_gdf[jeffco_index],'Voters=',Voters[i])

co_counties_gdf.to_file("co_counties.geojson", driver='GeoJSON')

co_counties_gdf.head(10)# nvoters = same for all precinct polygons
```

```python
WellsInBasins = gpd.sjoin(wells_gdf,riverbasins_gdf.to_crs(wells_gdf.crs), how='inner', predicate='intersects')

WellsInBasins['Wells']=1

WellsInBasins.head(10)
```

```python
WellsInBasins.to_crs(co_counties_gdf.crs)

WellsInBasins = WellsInBasins.groupby('HU6NAME').sum()

WellsInBasins
```

```python
#Get #wells for basins which have wells. 

Wells = WellsInBasins['Wells'].tolist()

for i in range(len(Wells)):
#    VotersInPrecincts.index[i],Voters[i])
    basin = WellsInBasins.index[i]
    
    basin_index = riverbasins_gdf[(riverbasins_gdf['HU6NAME'])==basin].index[:]
#    print('JeffCo Index:',len(jeffco_index),jeffco_index)
#    print(prec,jeffco_precincts_gdf[(jeffco_precincts_gdf['COUNTYPREC'])==prec])
    for j in range(0,len(basin_index)):
#        jeffco_precincts_gdf['Voters'][jeffco_index[j]]=Voters[i]
        riverbasins_gdf.loc[basin_index[j],'Wells']=Wells[i]
#        print(prec,j,Voters[i],jeffco_precincts_gdf['Voters'][j],jeffco_index)
#        print(VotersInprecincts.index[i],jeffco_index,jeffco_precincts_gdf[jeffco_index],'Voters=',Voters[i])

riverbasins_gdf.to_file("riverbasins.geojson", driver='GeoJSON')

riverbasins_gdf.head(10)# nvoters = same for all precinct polygons
```

```python
bmax = max(riverbasins_gdf['Wells'])
bmin = min(riverbasins_gdf['Wells'])

bmin, bmax
```

```python
#verify no missing/Nan
riverbasins_gdf.fillna(0.,inplace=True)

riverbasins_gdf.isnull().sum().sum()
```

```python
#get lat-lon of wells by converting to co_counties.gdf.crs on the fly

wells_gdf['lon'] = wells_gdf.to_crs(co_counties_gdf.crs)['geometry'].x
wells_gdf['lat'] = wells_gdf.to_crs(co_counties_gdf.crs)['geometry'].y

wells_gdf.head()
```

```python
#try creating a new wells_gdf based on lat-lon

#first, drop the geometry

wells_df = pd.DataFrame(wells_gdf.drop(columns='geometry'))

wells_df.head()
```

```python
#create the data frame
wells_gdf = gpd.GeoDataFrame(wells_df, geometry=gpd.points_from_xy(wells_df.lon,wells_df.lat))

wells_gdf.head()
```

```python
wells_gdf.set_crs(co_counties_gdf.crs)

wells_gdf.crs
```

```python
# create list of wells lat-lon

wells_latlon = []

for i in wells_gdf.index:
    point = []
    point.append(wells_gdf.at[i,'lat'])
    point.append(wells_gdf.at[i,'lon'])
    wells_latlon.append(point)
    
#print(wells_latlon)
```

```python
#make water map
water_map = fl.Map(location=[co_counties_gdf.centroid.y.mean(),
                              co_counties_gdf.centroid.x.mean()],
                zoom_start=7,tiles=None)
fl.TileLayer('cartodbpositron',name='BackGround',control=False).add_to(water_map)

colormap_counties = cm.linear.Set1_09.scale(0,cmax)

colormap_counties.caption = 'Number of Wells in County'
colormap_counties.add_to(water_map)

colormap_counties
```

```python
colormap_basins = cm.linear.Accent_08.scale(0,bmax)

colormap_basins.caption = 'Number of Wells in Basin'
colormap_basins.add_to(water_map)

colormap_basins
```

```python
# Add county data with mouseover
fl.GeoJson(co_counties_gdf, name='Counties',

           style_function = lambda x: {"weight":0.5, 
                                       'color':'grey',
#                                       'fillColor':'red',
                            'fillColor':colormap_counties(x['properties']['Wells']), 
                            'fillOpacity':0.5
               },
           
               highlight_function = lambda x: {'fillColor': '#000000', 
                                'color':'#000000', 
                                'fillOpacity': 0.25, 
                                'weight': 0.1
               },
               
               tooltip=fl.GeoJsonTooltip(
                   fields=['LABEL','Wells'],
                   aliases=['County:','Wells:'],
                   labels=True,
                   localize=True
               ),
               ).add_to(water_map)


```

```python
# Add wells
# Set max zoom level for clusters, also max zoom for spiderfy
zoom_max = 15

# Add Most of the Ponds and Pits to a MarkerCluster layer (clickable)
wells_layer = FastMarkerCluster(wells_latlon,wells_latname="Well Locations",overlay=True,control=True,disableClusteringAtZoom=zoom_max
                              )

wells_gdf.apply(lambda row:
                        fl.CircleMarker(name='Wells',
                                  location=[row['geometry'].y, row['geometry'].x],
                                  radius=0.1,
                                  color='grey',
                                  fill=True,
                                  fill_color='grey',
                                  fillOpacity=0.25,
                                  popup=row['WellName'],
                            
               highlight_function = lambda x: {'fillColor': '#000000', 
                                'color':'#000000', 
                                'fillOpacity': 0.25, 
                                'weight': 0.1
               },
               
               tooltip = "Well Name: %s<br>Well Depth: %s<br>Elevation: %s" % (row['WellName'], row['WLDepth'],row['WLElevatio'])             

                                 ).add_to(wells_layer), 
                         axis=1)

wells_layer.add_to(water_map)
```

```python
#create well_data, a heatmap

#well_data = [[row['lat'],row['lon']] for index,row in wells_gdf.iterrows()]

HeatMap(wells_latlon,name='Well Density',control=True,overlay=True).add_to(water_map)

HeatMap
```

```python
# Add river data with mouseover
fl.GeoJson(riverbasins_gdf, name='River Basins',
               style_function = lambda x: {"weight":1.0, 
                            'color':'blue',
#                            'fillColor':'transparent', 
                            'fillColor':colormap_basins(x['properties']['Wells']), 
                            'fillOpacity':0.5
               },
           
               highlight_function = lambda x: {'fillColor': '#000000', 
                                'color':'#000000', 
                                'fillOpacity': 0.25, 
                                'weight': 0.1
               },
               
               tooltip=fl.GeoJsonTooltip(
                   fields=['HU6NAME','Wells'],
                   aliases=['Basin:','Wells:'],
                   labels=True,
                   localize=True
               ),
               ).add_to(water_map)
```

```python
# Add control
fl.LayerControl(collapsed=False).add_to(water_map)


# Add minimap
MiniMap(tile_layer='openstreetmap',zoom_level_offset=-5).add_to(water_map)


water_map
```
